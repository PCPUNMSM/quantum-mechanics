# Movimiento armónico 1D

```python
%matplotlib inline
```

```python
import numpy
from numpy import linspace, zeros, sqrt, arctan2, sin, cos
from matplotlib import pyplot as plt
```

```python
def F(x, kx=1):
    return - kx * x
```

```python
m = 1 # kg
kx = sqrt(7) # N / m

N = 100000
to = 0
tf = 10
dt = (tf - to) / N
t = linspace(to, tf, N + 1)

x = zeros(t.shape)

vx = zeros(t.shape)

ax = zeros(t.shape)

x[0] = 0 # m

vx[0] = 1  # m/s
```

```python
for i, _ in enumerate(t[1:] , 1):
    vx[i] = vx[i-1] + ax[i-1] * dt
    x[i] = x[i-1] + vx[i] * dt
    ax[i] = F(x[i], kx) / m
```

## Trayectoria

```python
plt.plot(t, x)
```

## Energía

```python
plt.plot(
    t, 0.5 * m * vx * vx, 'r-',
    t, 0.5 * kx * x * x, 'g-.',
    t, 0.5 * m * vx * vx + 0.5 * kx * x * x, 'b--',
)
```
