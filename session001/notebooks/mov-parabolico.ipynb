# Movimiento parabólico

```python
%matplotlib inline
```

```python
from numpy import linspace, zeros, sqrt
from matplotlib import pyplot as plt
```

```python
m = 0.25

N = 1000
to = 0
tf = 12
dt = (tf - to) / N
t = linspace(to, tf, N + 1)

x = zeros(t.shape)
y = zeros(t.shape)

vx = zeros(t.shape)
vy = zeros(t.shape)

gx = 0
gy = -9.81

x[0] = 0 # m/s
y[0] = 0 # m/s

vx[0] = 25 # m/s
vy[0] = 50 # m/s
```

```python
# i = 0
# print("{0:5d}  {1:10.4f}  {2:10.4f}  {3:10.4f}  {4:10.4f}  {5:10.4f}"\
#     .format(i, time[i], x[i], y[i], vx[i], vy[i]))

for i, _ in enumerate(t[1:] , 1):
    vx[i] = vx[i-1] + gx * dt
    vy[i] = vy[i-1] + gy * dt
    
    x[i] = x[i-1] + vx[i] * dt
    y[i] = y[i-1] + vy[i] * dt
    
    # print("{0:5d}  {1:10.4f}  {2:10.4f}  {3:10.4f}  {4:10.4f}  {5:10.4f}"\
    #     .format(i, time[i], x[i], y[i], vx[i], vy[i]))
```

## Trayectoria

```python
plt.plot(x,y)
```

## Energía

```python
plt.plot(
    t, 0.5 * m * (vx * vx + vy * vy), 'r-',
    t, - m * gy * y, 'g-.',
    t, 0.5 * m * (vx * vx + vy * vy) - m * gy * y, 'b--',
)
```
